```python
import pandas as pd

# read input file
df_input_user_id = pd.read_excel('/Python Assignment.xlsx',sheet_name='(Input) User IDs')
#loc the actual position of table in dataframe
df1=df_input_user_id.iloc[10:31, 3:9]
#reset new dataframe inex
new_df_input_user_id=df1.reset_index(drop=True)
#rename column names according to use
new_df_input_user_id=new_df_input_user_id.rename(columns={'Unnamed: 3': 'S No', 'Unnamed: 4': 'Name','Unnamed: 5':'Team Name','Unnamed: 6':'User ID'})
new_df_input_user_id


```

```python
#do all steps hear as done privius one
df_ring=pd.read_excel('/Python Assignment.xlsx',sheet_name='(Input) Rigorbuilder RAW')
df2=df_ring.iloc[7:28, 2:7]
new_df_input_Rigorbuilder_RAW=df2.reset_index(drop=True)
new_df_input_Rigorbuilder_RAW=new_df_input_Rigorbuilder_RAW.rename(columns={'Unnamed: 2': 'S No', 'Unnamed: 3': 'Name','Unnamed: 4':'uid','Unnamed: 5':'total_statements','Unnamed: 6':'total_reasons'})

new_df_input_Rigorbuilder_RAW
```

```python
#merg both above 'new_df_input_user_id' and  'new_df_input_Rigorbuilder_RAW' data frame
merged_df = pd.merge(new_df_input_user_id, new_df_input_Rigorbuilder_RAW, on='Name')
#replace element 'Brandtech Lab'to 'BrandTech Lab' from column 'Team Name
merged_df['Team Name'] = merged_df['Team Name'].replace('Brandtech Lab', 'BrandTech Lab')
#rename 'Team Name' to 'Thinking Teams Leaderboard'
merged_df=merged_df.rename(columns={'Team Name':'Thinking Teams Leaderboard'})
merged_df
```

```python
#find mean with respect to 'total_statements' column and round it upto 2 decimal point
Average_Statements = round(merged_df.groupby('Thinking Teams Leaderboard')['total_statements'].mean(),2)
Average_Statements
```

```python
#find mean with respect to 'total_reasons' column and round it upto 2 decimal point
Average_Reasons=round(merged_df.groupby('Thinking Teams Leaderboard')['total_reasons'].mean(),2)
Average_Reasons
```

```python
#merg above two dataframe 'Average_Statements' and 'Average_Reasons' on basis of column-Team Name
Team_wise_leadarship1=pd.merge(Average_Statements, Average_Reasons, on='Thinking Teams Leaderboard')
#sort the value on decending order
Team_wise_leadarship=Team_wise_leadarship1.sort_values(by='total_statements', ascending=False)
#rename the column name according to need
Team_wise_leadarship=Team_wise_leadarship.rename(columns={'total_statements':'Average Statements','total_reasons':'Average Reasons'}) 

Team_wise_leadarship
```

```python
#creat new colum by adding column 'Average Statements' and 'Average Reasons'
Team_wise_leadarship['sum1']=Team_wise_leadarship['Average Statements']+Team_wise_leadarship['Average Reasons']

# Use the rank method to assign ranks to the values in the 'sum1' column
Team_wise_leadarship['rank'] = Team_wise_leadarship['sum1'].rank(method='min', ascending=False)


# Create a new column 'Team Rank' with the ranks

Team_wise_leadarship['Team Rank'] = Team_wise_leadarship['sum1'].rank(method='min', ascending=False)
Team_wise_leadarship['Team Rank']=Team_wise_leadarship['Team Rank'].astype(int)
# Drop the 'rank' column
Team_wise_leadarship = Team_wise_leadarship.drop('rank', axis=1)

#Drop the 'sum1' column
Team_wise_leadarship=Team_wise_leadarship.drop('sum1',axis=1)



# Reorder the columns of the dataframe
Team_wise_leadarship



```

```python
#assigning new dataframe
Leaderboard_individual=new_df_input_Rigorbuilder_RAW
#rename column name as we needed

Leaderboard_individual=Leaderboard_individual.rename(columns={'uid':'UID','total_statements':'No. of Statements','total_reasons':'No. of Reasons'})

#creat new column having sum of column 'No. of Statements' and 'No. of Reasons'
Leaderboard_individual['sum_value']=Leaderboard_individual['No. of Statements']+Leaderboard_individual['No. of Reasons']

#drop 'SNo' column
Leaderboard_individual = Leaderboard_individual.drop('S No', axis=1)

#sort dataframe according to decresing order of column 'No. of Statements, values
Leaderboard_individual = Leaderboard_individual.sort_values(by='sum_value', ascending=False)


# Create a new column 'Rank' with the ranks

Leaderboard_individual['Rank'] = Leaderboard_individual['sum_value'].rank(method='min', ascending=False)
Leaderboard_individual['Rank']=Leaderboard_individual['Rank'].astype(int)
# Drop the 'sum_value' column
Leaderboard_individual = Leaderboard_individual.drop('sum_value', axis=1)


Leaderboard_individual
```
